```python

from sqlalchemy import create_engine

df_ghana = pd.DataFrame(data_ghana)
df_nigeria = pd.DataFrame(data_nigeria)

# SQLite engine (file will be stored in Colab)
conn = create_engine("sqlite:///mydata.db")

# Save dataframe to SQL
df_ghana.to_sql("vaccine_data", conn, if_exists="replace", index=False)
df_nigeria.to_sql("vaccine_data", conn, if_exists="append", index=False)


print("Print Vaccine data stored in SQLite")
```
